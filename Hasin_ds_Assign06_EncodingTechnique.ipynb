# Label Encoding

```python
from pandas.api.types import is_numeric_dtype #Library Import
from pandas.api.types import is_string_dtype
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()  # Assign
for col in df_label.columns:    #loop over each Non Numeric Column
        if is_string_dtype(df_label[col]):
            df_label[col]=le.fit_transform(df_label[col])  #Fil Label Encoding
```

# One Hot Encoding

```python
from pandas.api.types import is_numeric_dtype  #Library Import
from pandas.api.types import is_string_dtype
from sklearn.preprocessing import OneHotEncoder
```

```python

for col in df_oneHot.columns:         #loop over each Non Numeric Column
    if is_string_dtype(df_oneHot[col]):
        dummy=pd.get_dummies(df_oneHot[col],drop_first=True,prefix=col) # Creat Dummy Table  and Delete First Column
        df_oneHot.drop(df_oneHot[[col]],axis=1, inplace=True)
        df_oneHot=pd.concat([df_oneHot,dummy],axis=1)
   
#new_hot=pd.concat([df_oneHot,dff],axis=1)
        
```

# Ordinal Encoding

```python
from pandas.api.types import is_string_dtype
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
nnumeric_col_list=[]                 # List of Non Numeric Column
for col in df_ordinal.columns:
    if is_string_dtype(df_ordinal[col]):
        nnumeric_col_list.append(col)  # Append Non Numeric column in the col_list               
for col in nnumeric_col_list: 
    unique=df_ordinal[col].unique()         # Unique Rows of Each Column
    ordinal=OrdinalEncoder(categories=[unique])  
    encoded=ordinal.fit_transform(df_ordinal[[col]]) # encoded each column
    new_col=pd.DataFrame(encoded,columns=[col])      #  Make new column 
    df_ordinal[col]=new_col                    #Replace column
   
```
