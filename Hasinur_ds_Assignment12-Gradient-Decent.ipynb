```python
import matplotlib.pyplot as plt
import numpy as np

n=len(X_true)
def mean_squared_error_func(Y_true,Y_pred):
    cost=sum((Y_true-Y_pred)**2/n)
    return cost

def gradient_Function():
    iteration=2000
    learning_rate=0.0001
    current_weight=0
    current_intercept=0
    stopping_threshold = 1e-6
    previous_cost=None
    
    costs=[]
    weights=[]
    
    for i in range(iteration):
        Y_pred=(current_weight*X_true)+current_intercept
        current_cost=mean_squared_error_func(Y_true,Y_pred)
        if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
            break
        previous_cost=current_cost
        
        costs.append(current_cost)
        weights.append(current_weight)
        
        #Calculating Gradient
        weight_derivative=(-2/n)*sum(X_true*(Y_true-Y_pred))
        intercept_derivative = -(2/n) * sum(Y_true-Y_pred)
        
        current_weight=current_weight-(learning_rate*weight_derivative)
        current_intercept=current_intercept-(learning_rate*intercept_derivative)
        print(f'Iteration {i}: Cost Function {current_cost} Weight {current_weight} Intercept {current_intercept}')
    plt.figure(figsize=(10,12))
    plt.plot(weights,costs)
    plt.scatter(weights,costs,marker='o',color='red')
    plt.title("Cost vs Weights")
    plt.ylabel("Cost")
    plt.xlabel("Weight")
    plt.show()
    
    return current_weight,current_intercept

def main():
    estimated_weight,estimated_intercerpt=gradient_Function()
    print(f"Estimated Weight: {estimated_weight} Estimated Intercept: {estimated_intercerpt}")
    
    Y_pred=(estimated_weight*29.070	)+estimated_intercerpt
    #plt.figure(figsize = (8,6))
    #plt.scatter(X, Y, marker='o', color='red')
   # plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='blue',markerfacecolor='red',
            # markersize=10,linestyle='dashed')
   # plt.xlabel("X")
   # plt.ylabel("Y")
   # plt.show()
    print(Y_pred)
 
if __name__=="__main__":
    main()    
```
